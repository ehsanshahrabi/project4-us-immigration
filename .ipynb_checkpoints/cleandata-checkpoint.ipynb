```python
# Convert the "Year" column to int64 data type in both DataFrames
inmigration_mexico.loc[:, "Year"] = inmigration_mexico["Year"].astype(int)
regressors_country_mexico.loc[:, "Year"] = regressors_country_mexico["Year"].astype(int)


# Merge the two DataFrames on the "Year" column
df_ml_mexico = pd.merge(inmigration_mexico, regressors_country_mexico, on="Year", how="inner")
df_ml_mexico
```

```python
df_ml_mexico_columns = df_ml_mexico.columns
df_ml_mexico_columns
```

```python
df_ml_mexico.info()
```

```python
import numpy as np

# Convert the columns to numeric data types
df_ml_mexico['Population_total'] = df_ml_mexico['Population_total'].astype(float)
df_ml_mexico['GDP_per_capita'] = df_ml_mexico['GDP_per_capita'].astype(float)

# Replace non-numeric values with NaN in the 'UR' column
df_ml_mexico['UR'] = pd.to_numeric(df_ml_mexico['UR'], errors='coerce')

# Convert the 'Gini_index' column to a numeric data type
df_ml_mexico['Gini_index'] = pd.to_numeric(df_ml_mexico['Gini_index'], errors='coerce')

# Use forward fill to impute missing values in the regressor columns
df_ml_mexico.fillna(method='ffill', inplace=True)

df_ml_mexico
```

```python
from prophet import Prophet

# Create a new DataFrame with only the columns needed for prediction
data_for_prophet = df_ml_mexico[['Year', 'Total_inmigrants', 'California', 'Population_total', 'GDP_per_capita', 'UR', 'Urban_population', 'Birth_rate_crude', 'Gini_index']].rename(columns={'Year': 'ds', 'Total_inmigrants': 'y'})

# Convert the "ds" column to datetime format
data_for_prophet['ds'] = pd.to_datetime(data_for_prophet['ds'], format='%Y')

# Replace missing values in regressor columns with the last available value of each regressor column
regressor_columns = ['Population_total', 'GDP_per_capita', 'UR', 'Urban_population', 'Birth_rate_crude', 'Gini_index']
for col in regressor_columns:
    last_year_data = df_ml_mexico.dropna(subset=[col], how='any').tail(1)
    last_year_value = last_year_data.iloc[0][col]
    data_for_prophet[col].fillna(last_year_value, inplace=True)

# Initialize and fit the model with regressors
model = Prophet()
for col in regressor_columns:
    model.add_regressor(col)
model.fit(data_for_prophet)

# Create a DataFrame with the future dates we want to predict (e.g., 2023)
future_dates = model.make_future_dataframe(periods=1, freq='Y')

# Change the format of the 'ds' column to integer (e.g., 2023)
future_dates['ds'] = future_dates['ds'].dt.year

# Add the regressor columns and set their values to the last available year value
for col in regressor_columns:
    future_dates[col] = last_year_value

# Make predictions for the future dates
forecast = model.predict(future_dates)

# Check if there are predictions for the year 2023
if (forecast['ds'] == 2023).any():
    prediction_2023 = forecast[forecast['ds'] == 2023]['yhat'].values[0]
    print("Predicted number of immigrants in California in 2023:", prediction_2023)
else:
    print("No predictions available for the year 2023.")
```

```python
#Predict without the regressors

# Create a new DataFrame with only the columns needed for prediction
data_for_prophet = df_ml_mexico[['Year', 'Total_inmigrants']].rename(columns={'Year': 'ds', 'Total_inmigrants': 'y'})

# Initialize and fit the model
model = Prophet()
model.fit(data_for_prophet)

# Create a DataFrame with the future dates we want to predict (e.g., 2023)
future_dates = model.make_future_dataframe(periods=1, freq='Y')

# Make predictions for the future dates
forecast = model.predict(future_dates)

# Extract the predicted value for 2023
if '2023-01-01' in forecast['ds'].astype(str).values:
    prediction_2023 = forecast[forecast['ds'] == '2023-01-01']['yhat'].values[0]
    print("Predicted number of immigrants in California in 2023:", prediction_2023)
else:
    print("No predictions available for the year 2023.")
```
