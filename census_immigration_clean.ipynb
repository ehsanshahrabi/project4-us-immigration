```
import numpy as np
import pandas as pd
import datetime as dt
```

```
immigration_csv = "Resources/immigration_data_2005_2021.csv"
```

```
immigration = pd.read_csv(immigration_csv)
immigration
```

```
# Create list of columns to drop
drop_columns = ['U.S. possessions', 'U.S. Dependencies', 'Guam', 'Puerto Rico', 'U.S. Armed Services posts']

# Drop the specified columns and create a new DataFrame
immigration_clean = immigration.drop(columns=drop_columns)

immigration_clean
```

```
# Path to the CSV file
census_data = "Resources/census_info.csv"

# Read the CSV file and create a Pandas DataFrame
census_df = pd.read_csv(census_data)

# Print
census_df
```

```
# Find columns that contain "MOE" in their names
columns_to_drop = [col for col in census_df.columns if 'MOE' in col]

# Drop the columns with "MOE" in their names
df_filtered = census_df.drop(columns=columns_to_drop)

df_filtered
```

```
transposed_df = df_filtered.transpose()
transposed_df
```

```
transposed_df.to_csv('census_clean.csv')
transposed_df.to_csv('Resources/census_clean.csv')
```
